```python
#Reading the data
df = pd.read_csv('ETH.csv')
#df = pd.read_csv('BNB-USD.csv')


#Showing the data
df

```

```python
df.dropna(inplace=True)
rows_with_NaN = df[df.isnull().any(axis=1)]
rows_with_NaN
```

```python
#Create a variable for prediction 'n' days out into the future
projection = 5

#Create a new column called prediction
df['Prediction'] = df[['Close']].shift(-projection)

#Show the dataset
df
```

```python
X = df
print(X)
```

```python
#Create the independent data set (X)
X = df.drop('Prediction', axis = 1)
X = df.drop('Date', axis = 1)
#Remove the last 14 rows
X = X[: -projection]
print(X)
```

```python
#Create the dependent dataset
y = df['Prediction'].values
y = y[: -projection]
print(y)
```

```python
#Split the data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,y,test_size = 0.25)
```

```python
from sklearn.ensemble import RandomForestRegressor
model1=RandomForestRegressor(n_estimators=10)
model1.fit(x_train, y_train)
```

```python
RFR_confidence = model1.score(x_test, y_test)
print('Random Forest Confidence: ', RFR_confidence)
```
